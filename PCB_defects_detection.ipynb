### Moving Labels and Images to YOLO Directory

```python
# ### Function: to_v5_directories
# This function organizes the dataset for YOLOv5 training by splitting images and labels into training and validation directories.
#
# Parameters:
# - `images_train_path` (str): Path to the directory where training images will be stored.
# - `images_val_path` (str): Path to the directory where validation images will be stored.
# - `labels_train_path` (str): Path to the directory where training labels will be stored.
# - `labels_val_path` (str): Path to the directory where validation labels will be stored.
# - `dataset_source` (str): Path to the source directory containing images and labels.
#
# The function:
# 1. Reads the images and labels from the source directory.
# 2. Splits them into training and validation sets based on a predefined ratio (80% training, 20% validation).
# 3. Copies the files to respective directories for YOLO training.

import os
from random import choice
import shutil

import os
import shutil
from random import choice

def to_v5_directories(images_train_path, images_val_path, labels_train_path, labels_val_path, dataset_source):
    """
    Split images and labels into training and validation sets, and move them to respective directories.

    Parameters:
    - images_train_path (str): Path where training images will be moved.
    - images_val_path (str): Path where validation images will be moved.
    - labels_train_path (str): Path where training labels will be moved.
    - labels_val_path (str): Path where validation labels will be moved.
    - dataset_source (str): Source directory containing the images and labels.

    Returns:
    - None
    """
    imgs = []  # List to store image filenames
    xmls = []  # List to store label filenames
    train_ratio = 0.8  # Ratio of images to be used for training
    val_ratio = 0.2  # Ratio of images to be used for validation

    # Listing all images and annotations in the dataset source directory
    for (dirname, dirs, files) in os.walk(dataset_source):
        for filename in files:
            if filename.endswith('.txt'):  # Labels are in .txt files
                xmls.append(filename)
            else:  # Images are not .txt files
                imgs.append(filename)

    # Determine number of images for training and validation
    countForTrain = int(len(imgs) * train_ratio)
    countForVal = int(len(imgs) * val_ratio)

    # Move images and labels to respective directories for training set
    for _ in range(countForTrain):
        fileJpg = choice(imgs)  # Randomly select an image
        fileXml = fileJpg[:-4] + '.txt'  # Corresponding label file
        shutil.copy(os.path.join(dataset_source, fileJpg), os.path.join(images_train_path, fileJpg))
        shutil.copy(os.path.join(dataset_source, fileXml), os.path.join(labels_train_path, fileXml))
        imgs.remove(fileJpg)
        xmls.remove(fileXml)

    # Move images and labels to respective directories for validation set
    for _ in range(countForVal):
        fileJpg = choice(imgs)  # Randomly select an image
        fileXml = fileJpg[:-4] + '.txt'  # Corresponding label file
        shutil.copy(os.path.join(dataset_source, fileJpg), os.path.join(images_val_path, fileJpg))
        shutil.copy(os.path.join(dataset_source, fileXml), os.path.join(labels_val_path, fileXml))
        imgs.remove(fileJpg)
        xmls.remove(fileXml)

    print("Training images:", countForTrain)
    print("Validation images:", countForVal)

    # Optionally, uncomment the line below to move the entire dataset folder after the operation
    # shutil.move(dataset_source, images_val_path)
```
